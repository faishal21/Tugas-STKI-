```python
#Lowercase

#kalimat = "Berikut ini adalah 5 negara dengan pendidikan terbaik di dunia adalah Korea Selatan, Jepang, Singapura, Hong Kong, dan Finlandia."
#lower_case = kalimat.lower()
#print(lower_case)

kalimat = "Saya Mahsiswa Udinus Semester 5 Fakultas Teknik Informatika. Saya Sedang Belajar. \n\n"
lower_case = kalimat.lower()
print (lower_case)
```

```python
#Removing Number

import re

#kalimat = "Berikut ini adalah 5 negara dengan pendidikan terbaik di dunia adalah Korea Selatan, Jepang, Singapura, Hong Kong, dan Finlandia."
#hasil = re.sub(r"\d+", "", kalimat)
#hasil

hasil = re.sub(r"\d+", "", lower_case)
print (hasil)
```

```python
#Removing Punctuation

import string

#kalimat = "Ini &adalah [contoh] kalimat? {dengan} tanda. baca?!!"
#hasil = kalimat.translate(str.maketrans("","",string.punctuation))
#hasil

hasil = hasil.translate(str.maketrans("","",string.punctuation))
print (hasil)
```

```python
#Removing Whitespace

#kalimat = " \t    ini kalimat contoh \t   \n"
#hasil = kalimat.strip()
#hasil

hasil = hasil.strip()
print (hasil)
```

```python
#kalimat = "rumah idaman adalah rumah yang bersih"
#pisah = kalimat.split()
#pisah

pisah = hasil.split()
print (pisah)
```
